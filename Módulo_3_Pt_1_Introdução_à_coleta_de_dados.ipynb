###API
Processo em 3 etapas:
1. Fazer uma requisição para a API: apontar uma URL base e parâmetros/filtros da requisição de dados
2. Obter uma resposta da API: resultado bem/mal sucedido, conteúdo/corpo
3. Tranformar a resposta em um objeto da linguagem: de XML/JSON/etc para uma tabela de dados 


####coletando dados do BC via API
1. Consultar a tabela desejada no SGS
2. Para saber o formato da planilha do SGS, entramos no portal https://dadosabertos.bcb.gov.br/organization/depec
e buscamos pelo nome da tabela (que descobrimos no SGS).

  Uma vez consultando a planilha nos dados abertos, saberemos o URL base a ser utilizado
3. Se clicarmos na URL, estaremos fazendo a requisição e obteremos a resposta
4. Além da URL base, podemos fazer requisições customizadas, através do uso dos parâmetros da API, que também estão indicadas na página de informação da base de dados

Ex. no caso de planilhas sgs
https://api.bcb.gov.br/dados/serie/bcdata.sgs.codigo_serie/dados?formato=json&dataInicial={dataInicial}&dataFinal={dataFinal}

```python
#consultando API do BC, para a série 1207
##bibliotecas necessárias:
import requests
```

```python
##definir a URL da requisição.
##é o modelo de requisiçao do SGS, conforme explicado acima 
url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.1207/dados?formato=json"

##executar a requisição
resposta = requests.get(url = url)
resposta #se a repsosta for diferente de 200, provavelmente, houve algum erro
```

```python
#obtida a resposta, para trazê-la via DataFrame:
resposta.json()
```

```python
#passando para um formato dataframe
import pandas as pd
tabela = pd.DataFrame(resposta.json())
```

###Trabalhando com SQL

####Banco de dados SQL: 
É um arquivo, que geralmente tem extensão ".db", no qual podemos armazenar dados de forma estruturada, organizados em tabelas que são consultáveis, sem necessidade de carregar tud para a memória. 

####Cenário 1:
Temos vários arquivos de dados para analisar e desejamos armazenar os dados em um banco de dados SQL (por exemplo, porque são pesados demais)

###Cenário 2:
Já existe um banco de dados SQL e precisamos fazer consulta de dados de uma tabela no banco SQL

#### Cenário 1: criando um banco de dados SQL a partir de arquivos CSV

```python
#primeiro, fazemos o upload do arquivo para o Colab
import pandas as pd
dados = pd.read_csv(
    filepath_or_buffer = "https://aluno.analisemacro.com.br/download/45589/?tmstv=1672743605",
    sep = ";",
    encoding = "latin_1"
    ) 
dados
```

```python
#segundo, criamos um banco de dados SQL
## Há varios sistemas. Usaremos o sqlite
#bibliotecas
from sqlalchemy import create_engine, inspect
#create engine serve para criarmos uma representação de banco de dados, que é onde salvaremos. Tipo uma URL
banco_de_dados = create_engine("sqlite:///dados_vivi/votacao2022.sqlite")
banco_de_dados
```

```python
#terceiro, se conectando ao banco de dados:
conexao = banco_de_dados.connect()
conexao
#agora, temos um banco de dados criado, mas em branco
```

```python
#quarto, criaremos um tabela no nosso banco de dados sqlite a partir do csv que já temos ("dados")
dados.to_sql(name = "votacao_acre", con = conexao, index = False) #esse ultimo argumento faz com que o indice do DF nao vire uma coluna no banco de dados
```

#### Cenário 2: fazendo consultas a um banco de dados SQL

```python
#primeiro, indicaremos a localização de um banco SQL e nos conectaremos a ele
from sqlalchemy import create_engine, inspect
banco_de_dados = create_engine("sqlite:///dados_vivi/votacao2022.sqlite")
conexao = banco_de_dados.connect()
```

```python
#segundo, vamos inspecionar o banco de dados, entendendo quais tabelas existem nele
##vendo o nome das tabelas
inspecao = inspect(banco_de_dados)
inspecao.get_table_names()
```

```python
#ver informacoes das colunas de uma tabela
inspecao.get_columns(table_name="votacao_acre")
```

```python
#importar a tabela completa do banco SQL para o Python
pd.read_sql(sql = "votacao_acre", con = conexao)
```

```python
#fazer uma consulta ao SQL sem importar toda a tabela de lá. Ou seja, desejamos ler apenas uma parte da tabela para não ficar muito pesado
pd.read_sql(sql = "SELECT NM_VOTAVEL, QT_VOTOS, NM_MUNICIPIO, NR_SECAO FROM votacao_acre WHERE NM_MUNICIPIO == 'RIO BRANCO'")
```

```python
#ao final do trabalho, importante encerrar a conexao:
conexao.close()
```
