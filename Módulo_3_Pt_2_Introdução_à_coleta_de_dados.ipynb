###Introdução à Web Scraping
A técnica consiste na criação de um código para coletar um código disponível em uma página na Internet (HTML)

1. Acessar a página
2. Navegar na página 
3. Baixar e organizar dados de interesse através de códigos automatizáveis


Não é simples. Tente verificar se existe uma API, um link de download ou se os dados estão disponíveis em outro lugar na Internet

####Quando usar?
  * Não há APIs documentadas
  * Não há botão para baixar dados
  * Dados não estruturados para análise (textos)
  * É necessário automatizar a extração de dados
  * Copiar e colar a informação para um arquivo local é inviável

####Aplicações
  * Tabelas de dados em sites
  * Extração de links para arquivos (por exemplo, se quiser baixar todos os arquivos excel de um site)
  * Monitoramento e comparação de preços (por exemplo, de ação)
  * Extração de textos em páginas
  * Etc

####Desvantagens
  * É necessário entender de linguagens de desenvolvimento de sites
  * páginas na Internet mudam e o código pode parar de funcionar
  * o código pode ser bloqueado por configurações de segurança da página (por exemplo, no caso de acessos sucessivos)

####Tags importantes
No código de HTML (do qual extrairemos os dados), existem tags. É necessário saber o que são, para entender de onde extrairemos os dados

Os principais estão no ppt da aula.
Em nosso exemplo a seguir, extreiremos dados de uma tabela (tag "table") 

####Passos para o web scraping
1. Verificar se o site permite a extração de dados

2. O arquivo robots.txt: Verificar se os site permite o acessos por bots (scrapers, crawlers, spiders). As regras de permissão podem ser verificadas no arquivo "robots.txt", localizado, em geral, na raiz do site.
Nesses arquivos, há instruções do tipo chave-valor, dizendo quais partes do site é permitido/proibido fazer web scraping de dados. Caso não excontre, envie um email ao dono do site

Ex: 
http://wikipedia.org/robots.txt

https://www.google.com/robots.txt

No python, temos a biblioteca robotspy, que retornará True ou False

```python
#instalando a robotspy
!pip install robotspy
#importando o módulo que precisamos "RobotsParser"
from robots import RobotsParser as r
#bibliotecas adicionais
import requests
from bs4 import BeautifulSoup as bs #vamos usar para extrair elementos da pg
```

```python
#verificando permissoes via robots.txt
#a url é o caminho, dentro da página da wikipedia que desejamos acessar 
##por exemplo, wiki: https://en.wikipedia.org/wiki/Main_Page 
r.from_uri("http://wikipedia.org/robots.txt").can_fetch(useragent = "ana", url = "/wiki")
#google, e url a pg. principal do google 
r.from_uri("https://www.google.com/robots.txt").can_fetch(useragent = "ana", url = "")
#google, url de pesquisa (/search)
r.from_uri("https://www.google.com/robots.txt").can_fetch(useragent = "ana", url = "/search")
```

```python
#exemplo prático de webscraping ##tabela de uma pagina no wikipedia

#1º passo: verficiação de permisão da pg.
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"
r.from_uri(url).can_fetch(useragent = "ana", url = "")

#2º passo: ler a pg
pagina = requests.get(url)
pagina #se a resposta for [200] é pq o site está funcionando

#3º passo: transformar o objeto que desejamos extrair em BS4 
##para isso, pegaremos o código da pg. e identificaremos onde está o objeto desejado
pagina.content
#passando o conteúdo da pg. para a função bs e pedindo que leia como html e imprimir de forma mais friendly
conteudo = bs(pagina.content, "html")
conteudo

#4º passo: procurar o elemento desejado no conteúdo da pg
##no nosso caso é a tabela principal
#identificando elementos importantes da pg
conteudo.title #titulo
conteudo.p #paragrafos
conteudo.h1 #cabeçalhos
```

```python
#4º passo: procurar o elemento desejado no conteúdo da pg
##no nosso caso é a tabela principal
#identificando elementos importantes da pg
conteudo.title #titulo
conteudo.p #paragrafos
conteudo.h1 #cabeçalhos
#função para encontrar elementos
conteudo.find_all("h1") #cabeçalhos de nivel 1
conteudo.find_all("table") #tabelas
```

```python
#checando cada tabela do site pra ver qual queremos
#vendo o total de tabelas
tabelas = conteudo.find_all("table")
len(tabelas)
tabelas
```

```python
#capturando a tabela que desejamos
##para isso, precisamos identificar o atributo class da nossa tabela
##passa um ctl.f no codigo acima
##ou, no elemento da pg. que desejamos, clica com botão direito em "inspecionar" e identifica o atributo class
```

```python
#5º passo: filtrar tabela pelo atributo class
tabela_desejada = conteudo.find_all("table", class_ = "wikitable")
tabela_desejada
```

```python
#6ºpassa:
##identificado o elemento desejado, converter o cod. html para um DataFrame
import pandas as pd
pd.read_html(io = str(tabela_desejada))[0] 
###esse 0, no final é pq, como só existe no site 1 tabela "wikitable", o nosso DF está capturando esse elemento do conteudo 
```
